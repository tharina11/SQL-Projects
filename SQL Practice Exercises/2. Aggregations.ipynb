```python
import duckdb
import pandas as pd

prices = pd.read_csv("prices.csv")

def run_sql(query):
    return duckdb.query(query).to_df()
```

1. How many rows are in the prices table?

```python
run_sql("""
SELECT
    count(*)
FROM prices;
""")
```

2. What is the average closing price across all tickers?

```python
run_sql("""
SELECT ROUND(AVG(close), 2)
FROM prices;
""")
```

3. For each ticker, calculate:

- average close price

- total traded volume

```python
run_sql("""
SELECT 
ticker,
ROUND(AVG(close), 2) AS avg_close,
SUM(volume) AS total_volume
FROM prices
GROUP BY ticker;
""")
```

4. For each ticker, compute:

- highest high

- lowest low

- average daily volume

```python
run_sql("""
SELECT 
ticker,
MAX(high) AS avg_close,
MIN(low) AS lowest_low,
AVG(volume) AS average_daily_volume
FROM prices
GROUP BY ticker;
""")
```

5. For each date, calculate total volume traded across all tickers.

```python
run_sql("""
SELECT 
SUM(volume) AS total_volume
FROM prices
GROUP BY date;
""")
```

6. Calculate the average close price per ticker only for dates after 2024-03-01.

```python
run_sql("""
SELECT
ticker,
ROUND(AVG(close), 2) AS average_close
FROM prices
WHERE date > '2024-03-01'
GROUP BY ticker;
""")
```

7. Show tickers whose average close price is greater than 300.

```python
run_sql("""
SELECT
ticker,
ROUND(AVG(close), 2) AS average_close
FROM prices
GROUP BY ticker
HAVING average_close > 300;
""")
```

8. For each ticker, compute:

- avg close as avg_close

- max volume as max_volume

- order by avg_close descending.

```python
run_sql("""
SELECT
ticker,
ROUND(AVG(close), 2) AS average_close,
MAX(volume) AS max_volume
FROM prices
GROUP BY ticker
ORDER BY average_close DESC;
""")
```

10. Explain the difference between:

            WHERE volume > 2000000


and

            HAVING SUM(volume) > 2000000

The WHERE clause fliters data by rows; before the aggregation. In contrast, the HAVING clause filters data by groups; after the aggregation.
